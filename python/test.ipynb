# Numerical Experiment: sPCE in Public Investment Problem

## Setup

```python
%matplotlib inline
import numpy as np
from sPCE import PublicInvestProblem
import matplotlib.pyplot as plt
```

## Sample Codes and Some Sanity Checks

Initialize a public investment problem environment.

Set the values of initial probability of state H as `q_vals = [0.1, 0.2, 0.3]`

Set the cost of contribution `x_vals = [0.2, 1.2]`

Set the size of the randomized prescription set num_quant=3, i.e., for each x, it can play contribution w.p. 0, 0.5, or 1.

Set time horizon `T = 2`.

There is a tolerance digit for belief update. We use the default value `tol_dig = 8`, which means the belief update function will always round result by `round(result, digits=8)`.




```python
env = PublicInvestProblem(q_vals=[0.1, 0.2, 0.3], x_vals=[0.2, 1.2], num_quant=3, T=2)
```

The environment can do the backward recursion step by step, `t = 1, 0` in this case.

Or run the solver directly by `env.solver()`. 

There is an option `team`. If it is `True`, the solver will solve the team problem.

Otherwise, it will solve the game problem and try to find sPCE.

```python
env.solver(team=True)
```

The social welfare matches the MATLAB results we have. 

One may run `env.compare()` to obtain the results from both team and game problems.

You don't have to run `env.solver()` first in order to run this comparison, 
but if you do that, `env.compare()` will directly run the other solver of the opposite type to save time.

```python
env.compare()
```
